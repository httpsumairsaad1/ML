# **Advance: Chart-1: Sepal Length vs Sepal Width**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Assuming 'df' is your DataFrame and has the columns 'SepalLengthCm' and 'SepalWidthCm'

# Create a 2D histogram for all species combined
heatmap_data, x_edges, y_edges = np.histogram2d(
    df['SepalLengthCm'],
    df['SepalWidthCm'],
    bins=(10, 10)
)

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data.T, cmap='coolwarm', cbar=True,
            xticklabels=np.round(x_edges, 2),
            yticklabels=np.round(y_edges, 2))

plt.title("Heatmap for Sepal Length vs. Sepal Width (All Species)")
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

# **Data Scaling**

```python
x= df.drop ('Species', axis=1)
# x=df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y= df['Species']
print(x,y)
```

# **Data Splitting**

```python
x_train, x_test, y_train, y_test = train_test_split (x,y, test_size=0.3)
```

```python
print('Input (X) For Training  : ' ,x_train.value_counts().sum())
print('Output (Y) For Training : ' ,y_train.value_counts().sum())
print('Input (X) For Testing   : ' ,x_test.value_counts().sum())
print('Output (Y) For Testing  : ' ,y_test.value_counts().sum())
```

# **ML MODEL Implementation**

### **Model: 1**

```python
from sklearn.metrics import accuracy_score
model = LogisticRegression(random_state=42, max_iter=200)

model.fit(x_train, y_train)

# Predict on training and testing data
Y_train_pred = model.predict(x_train)
Y_test_pred = model.predict(x_test)

# Calculate accuracy for training and testing datasets
accuracy_train = accuracy_score(y_train, Y_train_pred)
accuracy_test = accuracy_score(y_test, Y_test_pred)

# Display the accuracies
print('Logistic Regression Model')
print(f"Training Accuracy: {accuracy_train:.2f}")
print(f"Testing Accuracy: {accuracy_test:.2f}")
```

### **Model: 2**

```python

dt_model = DecisionTreeClassifier(random_state=20, max_depth=None)

dt_model.fit(x_train, y_train)

# Predict on training and testing data
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)

# Calculate accuracy for training and testing datasets
accuracy_train = accuracy_score(y_train, y_train_pred)
accuracy_test = accuracy_score(y_test, y_test_pred)

# Display the accuracies
print('Decision Tree Model')
print(f"Training Accuracy: {accuracy_train:.2f}")
print(f"Testing Accuracy: {accuracy_test:.2f}")

# Scatter plot of predicted vs actual labels
plt.figure(figsize=(8, 6))
plt.scatter(range(len(y_test)), y_test, label='Actual', alpha=0.7)
plt.scatter(range(len(y_test_pred)), y_test_pred, label='Predicted', alpha=0.7)
plt.xlabel("Sample Index")
plt.ylabel("Class Label")
plt.title("Actual vs Predicted Labels")
plt.legend()
plt.show()
```

### **Model: 2 - Advance Graph**

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.tree import DecisionTreeClassifier

# Train the DecisionTree model
dt_model = DecisionTreeClassifier(random_state=20, max_depth=None)
dt_model.fit(x_train, y_train)

# Predict on training and testing data
y_train_pred = dt_model.predict(x_train)
y_test_pred = dt_model.predict(x_test)

# Calculate accuracy for training and testing datasets
accuracy_train = accuracy_score(y_train, y_train_pred)
accuracy_test = accuracy_score(y_test, y_test_pred)

# Display the accuracies
print('Decision Tree Model')
print(f"Training Accuracy: {accuracy_train:.2f}")
print(f"Testing Accuracy: {accuracy_test:.2f}")

# Heatmap of predicted vs actual labels
plt.figure(figsize=(8, 6))
conf_matrix = confusion_matrix(y_test, y_test_pred)
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=set(y_test), yticklabels=set(y_test))
plt.xlabel("Predicted Labels")
plt.ylabel("Actual Labels")
plt.title("Heatmap of Actual vs Predicted Labels")
plt.show()
```
